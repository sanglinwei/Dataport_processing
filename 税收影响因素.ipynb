```python
import pandas as pd
df = pd.read_csv('./tax.csv')
df.columns = ['year', 'T', 'X1', 'X2', 
             'X3', 'X4', 'X5', 'X6', 'X7', 'X8']
df.head()
```

```python
yr = df['year'].tolist()
TAX = df['T'].tolist()

import matplotlib.pyplot as plt
plt.plot(yr, TAX)
plt.xticks(yr,rotation=60)
plt.title('Tax Revenue during 1997-2017')
plt.xlabel('Year')
plt.ylabel('Tax Revenue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='whitegrid', context='notebook')
cols = ['T', 'X1', 'X2', 
             'X3', 'X4', 'X5', 'X6', 'X7', 'X8']
sns.pairplot(df[cols], size=2.5)
plt.show()
```

```python
## get correlation between these 9 (corrcoef --> matrix)
## visualize on heatmap
import numpy as np
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm,
                cbar=True,
                annot=True,
                square=True,
                fmt='.2f',
                annot_kws={'size': 12},
                yticklabels=cols,
                xticklabels=cols)
plt.show()
```
